### Funciones 

```python
def plot_feature(df, col_name, target):
   
    f, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15,5), dpi=90)
    
    df_target = df.groupby([target, col_name]).size().reset_index()
    df_target = df_target.rename(columns={0:"conteo"})
    df_target["porcentajes"] = df_target['conteo'] / df_target.groupby(target)['conteo'].transform('sum') * 100
      
    sns.countplot(df[col_name], order=sorted(df[col_name].unique()), color='#432371', saturation=1, ax=ax1)
    ax1.set_xlabel(col_name)
    ax1.set_ylabel('Count')
    plt.xticks(rotation = 90)
    
    
    sns.barplot(data=df_target, y="porcentajes", x=col_name, hue=target, ax=ax2)
    ax2.set_xlabel(col_name)
    ax2.set_ylabel('Distribución')
    plt.xticks(rotation = 90)
    plt.tight_layout()
    
    sns.boxplot(data=df_target, x=col_name, hue=target, ax=ax3)
    ax2.set_xlabel(col_name)
    ax2.set_ylabel('Distribución')
    plt.xticks(rotation = 90)
    plt.tight_layout()
```

## Estudio de correlación entre variables 

```python
corr = df.corr()
fig = plt.figure(figsize=(11,8))
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
ax1 = fig.add_subplot(111)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.set(style="white")
sns.heatmap(corr, mask=mask,cmap=cmap, square=True)
ax1.tick_params(labelsize=5)
```

```python
corr = df.corr(method = 'spearman')
plt.figure(figsize = (7, 5))
sns.heatmap(corr.abs(), vmin=0, vmax=0.5 )
plt.show()
```

Es algo complicado de observar exactamente cuáles son las variables más correlacioneadas de esta forma por la gran cantidad que tenemos. Pero sin ninguna duda que hay bastante correlación entre muchas de ellas a parte de las cuáles ya se eperaba gran correlación. Me refiero a contribución y número de polizas de seguros, claramente está directamente relaciona con la otra y tendrán que sacarse de los modelos para mejorar la capacidad de procesamiento. 

La mejor forma de análizar la correlación en este caso que tenemos muchas variables y la capacidad de los gráficos no es la mejor es una tabla de comparación de variables.  

```python
new_corr = corr.abs()
new_corr.loc[:,:] = np.tril(new_corr, k=-1) # below main lower triangle of an array
new_corr = new_corr.stack().to_frame('correlation').reset_index().sort_values(by='correlation', ascending=True)
```

He decidido eliminar las correlaciones superiores al 85% ya que estás hacen refencia a contratación y número de polizas. Al estar directamente relacionadas no tiene sentido estudiarlas en este análisis. 

```python
new_corr = new_corr[(new_corr['correlation'] >= 0.50) & (new_corr['correlation'] <= 0.85)]
```

```python
new_corr.sort_values(by = ['correlation'], ascending=False)
```
